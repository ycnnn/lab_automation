```python
import numpy as np
import io
import base64
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(100)
y = np.random.normal(size=(100,100))
```

```python
s = io.BytesIO()
fig = plt.figure()
ax = fig.add_axes((0,0,1,1))
fig.canvas.draw()
line = ax.plot(x,x)
for i in range(len(y)):
    line[0].set_ydata(y[1])
    ax.relim()
    fig.savefig(s, bbox_inches="tight")
```
